```python

#load csv files to dataframes
vaccine = pd.read_csv(path_vaccine)
covid = pd.read_csv(path_covid)
variant = pd.read_csv(path_variant)

variant.rename(columns = {'Code':'iso_code', 'Day':'date'}, inplace=True)

## Create separate year, month, and day columns
variant_date_split = variant['date'].str.split('-', expand=True)
variant['year'] = variant_date_split[0]
variant['month'] = variant_date_split[0]
variant['day'] = variant_date_split[2]

covid_date_split = covid['date'].str.split('-', expand=True)
covid['year'] = covid_date_split[0]
covid['month'] = covid_date_split[1]
covid['day'] = covid_date_split[2]

vaccine_date_split = vaccine['date'].str.split('-', expand=True)
vaccine['year'] = vaccine_date_split[0]
vaccine['month'] = vaccine_date_split[1]
vaccine['day'] = vaccine_date_split[2]

## Subset covid and vaccine data 
vaccine['iso_date'] = vaccine['iso_code'] + '_' + vaccine['date']
covid['iso_date'] = covid['iso_code'] + '_' + covid['date']
variant['iso_date'] = variant['iso_code'] + '_' + variant['date']

vaccine.set_index(['iso_date'], inplace = True, drop = False)
covid.set_index(['iso_date'], inplace = True, drop = False)
variant.set_index(['iso_date'], inplace = True, drop = False)


variant = variant[variant['year']=='2021']
intersection_index = variant.index.intersection(covid.index)
intersection_index = intersection_index.intersection(vaccine.index)

covid = covid[covid['iso_date'].isin(intersection_index)]
vaccine = vaccine[vaccine['iso_date'].isin(intersection_index)]

# # # #join data frames
non_dupes = vaccine.columns.difference(covid.columns)
full_biweekly = covid.join(vaccine[non_dupes], how='outer', lsuffix='_covid', rsuffix='_vaccine')
# non_dupes_all = variant.columns.difference(full.columns)
full_biweekly = full_biweekly.join(variant, how='outer', rsuffix='_variant')

full_biweekly.to_csv('full_biweekly.csv')

```
